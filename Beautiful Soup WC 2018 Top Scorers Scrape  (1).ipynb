## Beautiful Soup WC Top Scorers Practice Scrape
- I use beautiful soup and requests to do a simple scrape of the espn website which contains the list of all the top scorers of the 2018 world cup ! 

```python
# Import Libraries
import requests
from bs4 import BeautifulSoup 
```

```python
# Create a variable for the url which contains the information you are trying to scrape
url = 'http://www.espnfc.us/fifa-world-cup/4/statistics/scorers'
```

```python
#get() which retrieves the information from the link
ret = requests.get(url)
```

```python
#These two functions just allow me to confirm that we are working with an HTML file, that we have the right content and Response 200 (standard)
print(ret)
print(ret.content)
```

```python
# Parses out the content of the url we specify
soup = BeautifulSoup(ret.content, 'html.parser')
```

```python
#The prettify function makes html tags more "pretty" clear for analysis 
soup.prettify()
```

```python
# Find the tag you are looking for ( you can use inspect element on the actual page).
#In this case the players can all be found under the tbody tag. 
players_1 = soup.find_all('tbody')
```

```python
#Here I use len(tbody_list) to check how many tbody items are being returned, and we get 1 which is what we want, one set of data that contains all the top scorers !
#Since our len(tbody_list) only returns one value we don't need to specify the dictionary.
len(tbody_list)
```

```python
# I want to narrow this down more to include the names of players who scored in the world cup
# Please note that I used tbody_list[0] instead of only tbody_list.find_all because find_all gives you a list of results and the find_all function does not take a list as an attribute
player_name = players_1[0].find_all('a')
player_name
```

```python
# We now want to extract the text (use the .text) from this function to clean up all the messy html tags in the list
final_data = []
for i in player_name:
    final_data.append(i.text)
```

```python
# We now want to remove the countries from this list to only ha
final_data
```

```python
# A litte hack to loop over the previous list and only get the names 
scorers = []
for i in range(len(final_data)):
    if i % 2 == 0:
        scorers.append(final_data[i])
    
scorers
```

```python
# A pretty pandas dataframe with all the top scorers from the 2018 world cup :)
import pandas as pd
df = pd.DataFrame(scorers)
df
```
